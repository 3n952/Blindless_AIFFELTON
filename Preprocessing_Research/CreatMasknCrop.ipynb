```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
import random
import math
import mask
import sys

sys.path.append("/visuworks/Blindless_AIFFELTON/script")

import preprocess

SOURCE = '/visuworks/Dataset/Selected Dataset 2'
CLIPLIMIT = 3.0
GRIDSIZE = (8, 8)
THRESHOLD_VALUE = 4
```

```python
samples_imgs_list = os.listdir(os.path.join(SOURCE, "Images"))
num_images_to_select = 10
selected_images = random.sample(samples_imgs_list, num_images_to_select)

for i, image_path in enumerate(selected_images):
    image_path = os.path.join(SOURCE, "Images", image_path)
    image = cv2.imread(image_path)
    r, g, b = cv2.split(image)
        
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    applied_clahe = preprocess.apply_clahe(image, CLIPLIMIT, GRIDSIZE)
    plt.imshow(applied_clahe, cmap='gray') 
    plt.title('apply clahe')
    
    plt.subplot(1, 3, 3)
    black_mask = cv2.inRange(image, (0, 0, 0), (0, 0, 0))
    masked_image = image.copy()
    masked_image[black_mask != 0] = [255, 255, 255]
    plt.imshow(cv2.cvtColor(masked_image, cv2.COLOR_BGR2RGB))
    plt.title('Black Pixel Mask')

    plt.show()
```
